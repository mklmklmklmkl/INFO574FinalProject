```python
import pandas as pd
# import seaborn as sns
import numpy as np
```

```python
# Netflix_Dataset_Movie
df1 = pd.read_csv('Netflix_Dataset_Movie.csv')
df1["Name"] = df1["Name"].str.lower()

df1
```

```python
# netflix_titles_2
df2 = pd.read_csv('netflix_titles_2.csv')
df2["title"] = df2["title"].str.lower()

df2
```

```python
# Check if each value in df2["title"] is in df1["name"]

# df2['has_corresponding'] = df2['title'].isin(df1['name'])

corresponding_values = df2['title'].isin(df1['Name'])

# Count the number of True values
true_count = corresponding_values.sum()

print(true_count)
```
